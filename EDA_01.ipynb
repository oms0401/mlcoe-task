#ADULT VIEWERSHIP

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df_adults = df[df['Age'] > 18]
colo=['b','y','r']
df_country = df_adults['Country'].value_counts()
df_country.plot(kind='bar',
                figsize=(10, 6),
               color=colo,
               )
plt.title('Countries with 19+ viewers')
plt.xlabel('Country')
plt.ylabel('Number of viewers')
plt.show()
```

# plot countrywise normal peoples

```python
country_wise_viewers = df['Country'].unique()
no_of_viewers = df['Country'].value_counts()
print(country_wise_viewers)
print(no_of_viewers)
col=['r','g','b','y','m']
plt.figure(figsize=(10,10))
plt.bar(country_wise_viewers,
        no_of_viewers,
       color=col,
        edgecolor='black',
        alpha=0.5,
       )
plt.grid()
plt.title('Country wise comparision of viewers')
plt.xlabel('Coutry')
plt.ylabel('number of viewers')
```

# compare monthly revenue on basis of subscriptions

```python
sub = df['Subscription Type'].unique()
print(sub)
rev= df['Subscription Type'].value_counts()
print(rev)
c2 = ['b','m','g']
plt.bar(sub,
        rev,
       color = c2,
        edgecolor='black',
        alpha=0.5,
       )
plt.grid()
plt.xlabel('Subscription type')
plt.ylabel('Revenue')
plt.title('catagerywise revenue')
```

# Drawing pie chart on basis of country involved

```python
plt.pie(no_of_viewers,
       labels = country_wise_viewers,
       )
plt.title('countrywise share')
```
